```python
# separar ademas si la ultima palabra entre parentesis tambien calza
import re

primer_palabra = pd.Series(map(lambda x: str(x).split()[0], df["Nombre"]))
parentesis_palabra = pd.Series(map(lambda x: re.findall(r"\((.*?)\)", str(x)), df["Nombre"]))
primer_termino_parentesis = pd.Series(map(lambda x: str(x).split(), parentesis_palabra))

primer_termino_parentesis
```
